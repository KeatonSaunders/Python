```python
import fix_yahoo_finance as yf
from datetime import datetime, timezone, timedelta
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
```

```python
start_date = "2013-01-01"

start_date = datetime.strptime(start_date,'%Y-%m-%d')
end_date = datetime.now()    

df = yf.download("ATMA.L", start=start_date, end=end_date)

df.head()
```

```python
df['MA50'] = df['Open'].rolling(50).mean()
df['MA200'] = df['Open'].rolling(200).mean()
df[['Open','MA50','MA200']].plot(label='df',figsize=(16,8));
```

```python
df['returns'] = df['Close'].pct_change(1)
df.head()
```

```python
df['returns'].hist(bins=50);
```

```python
df['Rolling stddev']=df['Close'].rolling(window=250).std()
df.head()
```
