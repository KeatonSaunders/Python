# Encoding categorical variables

````python
# Import LabelEncoder
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
````

````python
# Converting string labels into numbers.
newdf = pd.DataFrame()
for column in df.columns:
    if column not in ('Duration in months','Credit Amount','Installment rate % of disposable income','Present residence since','Age in years','Number of existing credit at the bank','Number of people being liable to provide maintenance for','Good/bad'):
        newdf[column]=le.fit_transform(df[column])
````

````python
df = newdf
df
````

# Standardize the Variables

````python
from sklearn.preprocessing import StandardScaler
````

````python
scaler = StandardScaler()
````

````python
scaler.fit(df.drop('Good/bad',axis=1))
````

````python
scaled_features = scaler.transform(df.drop('Good/bad',axis=1))
````

````python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat['Good/bad'] = df['Good/bad']
df_feat.head()
````

# Train test split

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df['Good/bad'],
                                                    test_size=0.30)
````

# Using KNN

````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
knn = KNeighborsClassifier(n_neighbors=1)
````

````python
knn.fit(X_train,y_train)
````

# Predictions and evaluations

````python
pred = knn.predict(X_test)
````

````python
from sklearn.metrics import classification_report,confusion_matrix
````

````python
print(confusion_matrix(y_test,pred))
````

````python
print(classification_report(y_test,pred))
````

# Choosing a K Value

````python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
````

````python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate');
````

# Retrain with new K value

````python
# NOW WITH K=12
knn = KNeighborsClassifier(n_neighbors=12)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=12')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
````
